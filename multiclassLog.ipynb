```python
from datasets import load_dataset
import numpy as np

# Load the dataset from Hugging Face
dataset = load_dataset("sst")
dataset.column_names
```

```python
import pandas as pd

train_df = pd.DataFrame(dataset["train"])
test_df = pd.DataFrame(dataset["test"])

# Save the train and test DataFrames to CSV files
train_df.to_csv("dataset.csv", index=False)
test_df.to_csv("test_dataset.csv", index=False)
# Assuming 'dataset' is a dictionary containing 'train' and 'test' keys, each containing the respective datasets
train_sentences = dataset["train"]["sentence"]
train_labels = dataset["train"]["label"]
test_sentences = dataset["test"]["sentence"]
test_labels = dataset["test"]["label"]

# Display the first few elements of each list
print("Train Sentences:", train_sentences[:5])
print("Train Labels:", train_labels[:5])
print("Test Sentences:", test_sentences[:5])
print("Test Labels:", test_labels[:5])
```

```python
def map_sentiment_score(score):
    if score <= 0.2:
        return 0  # Very negative
    elif score <= 0.4:
        return 1  # Negative
    elif score <= 0.6:
        return 2  # Neutral
    elif score <= 0.8:
        return 3  # Positive
    else:
        return 4  # Very positive
# Preprocess the dataset
def preprocess_data(sentences, labels):
    preprocessed_sentences = []
    preprocessed_labels = []
    for sentence, score in zip(sentences, labels):
        sentiment_class = map_sentiment_score(score)
        preprocessed_sentences.append(sentence)
        preprocessed_labels.append(sentiment_class) # Fixed the typo here
    return preprocessed_sentences, preprocessed_labels
preprocessed_train_sentences, preprocessed_train_labels = preprocess_data(train_sentences, train_labels)
preprocessed_test_sentences, preprocessed_test_labels = preprocess_data(test_sentences, test_labels)
```

```python


import numpy as np

def extract_bigrams(sentences):
    bigrams = set()
    for sentence in sentences:
        words = sentence.split()
        for i in range(len(words) - 1):
            bigram = ' '.join(words[i:i+2])
            bigrams.add(bigram)
    return list(bigrams)

def create_bigram_index(bigrams):
    bigram_index = {bigram: i for i, bigram in enumerate(bigrams)}
    return bigram_index

import numpy as np

def one_hot_encode_bigrams(sentences, bigrams):
    bigram_index = {bigram: i for i, bigram in enumerate(bigrams)}
    one_hot_encoded = np.zeros((len(sentences), len(bigrams)), dtype=int)
    for i, sentence in enumerate(sentences):
        words = sentence.split()
        for j in range(len(words) - 1):
            bigram = ' '.join(words[j:j+2])
            if bigram in bigram_index:
                index = bigram_index[bigram]
                one_hot_encoded[i, index] = 1
    return one_hot_encoded


# Extract bigrams from sentences
bigrams = extract_bigrams(preprocessed_train_sentences)

# One-hot encode bigrams in training sentences
X_train = one_hot_encode_bigrams(preprocessed_train_sentences, bigrams)

# One-hot encode bigrams in testing sentences
X_test = one_hot_encode_bigrams(preprocessed_test_sentences, bigrams)






```

```python
import numpy as np
import matplotlib.pyplot as plt

class MultiClassLogisticRegression:
    # ...
    def __init__(self, n_iter=10, thres=1e-3, lr=0.1):
        self.n_iter = n_iter
        self.thres = thres
        self.lr = lr
        self.weights = None

    # def add_bias(self, X):
    #   return np.insert(X, 0, 1, axis=1)

# Modify the fit method to add bias to X
    def fit(self, X, y, batch_size=16, rand_seed=4, verbose=False):
      np.random.seed(rand_seed)
      self.classes = np.unique(y)
      self.class_labels = {c: i for i, c in enumerate(self.classes)}
      X = self.add_bias(X)
      y = self.one_hot(y)
      self.loss = []
      self.weights = np.zeros(shape=(len(self.classes), X.shape[1]))
      self.fit_data(X, y, batch_size, verbose)
      return self

    def fit_data(self, X, y, batch_size, verbose):
        i = 0
        n_samples = X.shape[0]
        indices = np.arange(n_samples)
        while not self.n_iter or i < self.n_iter:
            print(i)
            self.loss.append(self.cross_entropy(y, self.predict_(X)))
            np.random.shuffle(indices)  # Shuffle indices
            for start_idx in range(0, n_samples, batch_size):
                batch_indices = indices[start_idx : start_idx + batch_size]
                X_batch, y_batch = X[batch_indices], y[batch_indices]
                error = y_batch - self.predict_(X_batch)
                update = self.lr * np.dot(error.T, X_batch)
                self.weights += update
            if np.abs(update).max() < self.thres:
                break
            if i % 1000 == 0 and verbose:
                print(f'Training Accuracy at {i} iterations is {self.evaluate_(X, y)}')
            i += 1

    def predict(self, X):
        return self.predict_(self.add_bias(X))

    def predict_(self, X):
        pre_vals = np.dot(X, self.weights.T).reshape(-1, len(self.classes))
        return self.softmax(pre_vals)

    def softmax(self, z):
        exp_vals = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_vals / np.sum(exp_vals, axis=1, keepdims=True)

    def predict_classes(self, X):



        self.probs_ = self.predict(X)
        return np.vectorize(lambda c: self.classes[c])(np.argmax(self.probs_, axis=1))

    def add_bias(self, X):
        if self.weights is None:
            return np.insert(X, 0, 1, axis=1)
        elif X.shape[1] == self.weights.shape[1]:
            return X
        else:
            return np.insert(X, 0, 1, axis=1)

    def one_hot(self, y):
        return np.eye(len(self.classes))[np.vectorize(lambda c: self.class_labels[c])(y).reshape(-1)]

    def score(self, X, y):
        return np.mean(self.predict_classes(X) == y)

    def evaluate_(self, X, y):
        return np.mean(np.argmax(self.predict_(X), axis=1) == np.argmax(y, axis=1))

    def cross_entropy(self, y, probs):
        return -1 * np.mean(y * np.log(probs))
```

```python
def confusion_matrix(y_true, y_pred, num_classes):
    # Initialize confusion matrix with zeros
    conf_matrix = np.zeros((num_classes, num_classes), dtype=int)

    # Fill the confusion matrix
    for true_label, pred_label in zip(y_true, y_pred):
        conf_matrix[true_label][pred_label] += 1

    return conf_matrix
```

```python
def generate_confusion_matrix(true_labels, predicted_labels, num_classes):
    confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)
    for true_label, predicted_label in zip(true_labels, predicted_labels):
        true_label = int(true_label[0]) if isinstance(true_label, np.ndarray) else int(true_label)
        predicted_label = int(predicted_label[0]) if isinstance(predicted_label, np.ndarray) else int(predicted_label)
        confusion_matrix[true_label][predicted_label] += 1
    return confusion_matrix

def compute_precision_recall_f1(confusion_matrix):
    num_classes = len(confusion_matrix)
    precision = np.zeros(num_classes)
    recall = np.zeros(num_classes)
    f1_score = np.zeros(num_classes)
    for i in range(num_classes):
        true_positive = confusion_matrix[i][i]
        false_positive = sum(confusion_matrix[:, i]) - true_positive
        false_negative = sum(confusion_matrix[i, :]) - true_positive
        precision[i] = true_positive / (true_positive + false_positive) if true_positive + false_positive > 0 else 0
        recall[i] = true_positive / (true_positive + false_negative) if true_positive + false_negative > 0 else 0
        f1_score[i] = 2 * precision[i] * recall[i] / (precision[i] + recall[i]) if precision[i] + recall[i] > 0 else 0
    return precision, recall, f1_score

def macro_averaged_precision_recall_f1(precision, recall, f1_score):
    macro_precision = np.mean(precision)
    macro_recall = np.mean(recall)
    macro_f1_score = np.mean(f1_score)
    return macro_precision, macro_recall, macro_f1_score
```
